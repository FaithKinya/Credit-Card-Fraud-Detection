Importing the Dependancies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Preprocessing

```python
credit_df = pd.read_csv("C://Users//Administrator//Documents//My Datasets//creditcard.csv (1)//creditcard.csv")
```

```python
credit_df.info()
```

```python
# check the missing values
credit_df.isnull().sum()
```

```python
credit_df['Class'].value_counts()
```

The dataset is highly unbalanced

0 --> means not fraud

1 --> means fraud

```python
# separating the data for analysis
legit = credit_df[credit_df.Class == 0]
fraud = credit_df[credit_df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of legit transactions in the Amount column
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_df.groupby('Class').mean()
```

Under-Sampling

Build a Sample Dataset Containing Similar Distribution of normal transaction and fradulent transaction

Number of Fradulent Transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

Concatenating Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
X = new_dataset.drop(['Class'], axis=1)
Y = new_dataset['Class']
```

```python
print(X.head())
```

```python
print(Y)
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```
